# Regression in PyTorch

```python
import torch
import matplotlib.pyplot as plt

x = torch.tensor([0, 1, 2, 3, 4, 5, 6, 7.]) # E.g.: Dosage of drug for treating Alzheimer's disease
x
```

```python
#y = -0.5*x + 2 + torch.normal(mean=torch.zeros(8), std=0.2) # standard deviation of the noise
# y = mx + b, y foram criados com essa equation.
```

```python
y = torch.tensor([1.86, 1.31, .62, .33, .09, -.67, -1.23, -1.37]) # E.g.: Patient's "forgetfulness score"
y # valores criados com a funo comentada acima
```

```python
fig, ax = plt.subplots()
plt.title("Clinical Trial")
plt.xlabel("Drug dosage (mL)")
plt.ylabel("Forgetfulness")
_ = ax.scatter(x, y)
```

Iniciamos o parametro m da slope com um valor "random" de 0.9

```python
m = torch.tensor([0.9]).requires_grad_()
m
```

```python
# o mesmo pro y-intercept b
b = torch.tensor([0.1]).requires_grad_()
b
```

```python
def regression(my_x, my_m, my_b):
    return my_m*my_x + my_b
```

```python
def regression_plot(my_x, my_y, my_m, my_b):

    fig, ax = plt.subplots()

    ax.scatter(my_x, my_y)

    x_min, x_max = ax.get_xlim()
    y_min = regression(x_min, my_m, my_b).detach().item() # fitting uma linha  data
    y_max = regression(x_max, my_m, my_b).detach().item()

    ax.set_xlim([x_min, x_max])
    _ = ax.plot([x_min, x_max], [y_min, y_max])
```

```python
regression_plot(x, y, m, b)
```

## Machine Learning

1. **Forward Pass**

```python
yhat = regression(x, m, b)
yhat
```

2. **Comparar yhat com y para calcular custo C**

H um mtodo MSELoss, mas vamos definir um prprio.

image.png

soma todas as diferenas ao quadrado, divide para pegar a mdia

```python
def mse(my_yhat, my_y):
    sigma = torch.sum((my_yhat - my_y)**2)
    return sigma/len(my_y)
```

```python
C = mse(yhat, y)
C
```

3. **Usar autodiff para calcular gradiente de C em relao aos parmetros**

```python
C.backward() # diferencia at os tensors mais internos, C  o mais externo
```

```python
m.grad # reduzindo o m, reduz MUITO mais o custo
```

```python
b.grad # reduzindo o b, reduz um pouco custo
```

4. Gradient Descent

```python
optimizer = torch.optim.SGD([m,b], lr=0.01)
# lista dos parametros a serem ajustados
# learning rate
```

```python
optimizer.step()
```

```python
m, b
```

```python
# replottar
regression_plot(x, y, m, b)
```

```python
C = mse(regression(x, m, b), y)
C
```

Loop denovo

```python
epochs = 1000
for epoch in range(epochs):
  optimizer.zero_grad() # reseta grads pra 0

  yhat = regression(x,m,b) # 1
  C = mse(yhat,y) # 2

  C.backward() # 3
  optimizer.step() # 4

  print('Epoch {}, cost {}, m grad {}, b grad {}'.format(epoch, '%.3g' % C.item(), '%.3g' % m.grad.item(), '%.3g' % b.grad.item()))
```

```python
regression_plot(x, y, m, b)
```
