## Frobenius Norm

```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import tensorflow as tf
```

```python
X = np.array([[1,2],[3,4]])
X
```

```python
(1**2 + 2**2 + 3**2 + 4**2)**(1/2)
```

```python
np.linalg.norm(X)
```

```python
X_pt = torch.tensor([[1,2],[3,4.]])
X_pt
```

```python
torch.norm(X_pt)
```

```python
X_tf = tf.Variable([[1,2],[3,4.]])
X_tf
```

```python
tf.norm(X_tf)
```

## Matrix Multiplication

```python
A= np.array([[3,4],[5,6],[7,8]])
A
```

```python
b = np.array([1,2])
b
```

```python
np.dot(A,b)
```

```python
A_pt = torch.tensor([[3,4],[5,6],[7,8.]])
A_pt
```

```python
b_pt = torch.tensor([1,2.])
b_pt
```

```python
torch.matmul(A_pt,b_pt)
```

```python
A_tf = tf.Variable([[3,4],[5,6],[7,8.]])
A_tf
```

```python
b_tf = tf.Variable([1,2.])
b_tf
```

```python
tf.linalg.matvec(A_tf,b_tf)
```

### Matrix x Matrix

```python
B = np.array([[1,9],[2,0]])
B,A
```

```python
np.dot(A,B)
```

Not commutative (AB =/= BA)

```python
B_pt = torch.from_numpy(B) #cleaner conversion
B_pt
```

```python
B_pt = torch.tensor([[1,2],[9,0.]]).T #creating the same tensor with transposition
B_pt
```

```python
torch.matmul(A_pt,B_pt)
```

```python
B_tf = tf.convert_to_tensor(B,dtype=tf.float32)
B_tf
```

```python
tf.matmul(A_tf,B_tf)
```

## Matrix Inversion

image.png

```python
X = np.array([[4,2],[-5,-3]])
X
```

```python
Xinv = np.linalg.inv(X)
Xinv
```

```python
y = np.array([4,-7])
y
```

```python
w = np.dot(Xinv,y)
w
```

```python
np.dot(X,w) # y = X.w
```

```python
torch.inverse(torch.tensor([[4,2],[-5,-3.]]))
```

```python
tf.linalg.inv(tf.Variable([[4,2],[-5,-3.]]))
```
